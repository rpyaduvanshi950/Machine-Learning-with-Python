```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(0)
X=np.random.randint(low=0, high=100, size=(100), dtype=int)
```

```python
X=np.array(X)
```

```python
Y=3*X+100+np.random.randint(low=0, high=20, size=(100), dtype=int)
```

```python
Y=np.array(Y)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def hypothesis(w,x,b):
    (w_X_b) = (w * x + b).reshape(-1)
    return (w_X_b)
    
```

```python
def loss_f(Y,h):
    return np.mean(np.subtract(Y,h)**3)
```

```python
def gradient_descent(w,b,X,Y,alpha):
    h=hypothesis(w,X,b)
    # Reshape arrays to ensure compatibility for element-wise operations
    dw = -3 * np.sum((X * (Y - h) ** 2* np.sign(Y-h) ))

    db = -3 * np.sum((Y - h) ** 2 * np.sign(Y - h))

    w=w-alpha*dw
    b=b-alpha*db
    return w,b
```

```python
epoch=5000000
alpha=0.000000001
```

```python
w = np.random.randn(1)
b = np.random.randn(1)
h=hypothesis(w,X,b)
l=[]
for i in range(epoch):
    w,b=gradient_descent(w,b,X,Y,alpha)
    l.append(loss_f(Y,h))
y_pred=w*X+b
print(b.shape)
plt.scatter(X, Y)
plt.plot(X, y_pred, 'r-', label='Predictions')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
print(w)
print(b)
```
