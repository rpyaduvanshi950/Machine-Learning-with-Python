# **Classification Using Decision Trees**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot


file=pd.read_csv("/content/wind_quality.txt")

data=pd.DataFrame(file)

data['value']=''


for i in range(len(data)):
  if data.at[i,'quality']>=7.0:
    data.at[i,'value']=1
  else:
    data.at[i,'value']=0

data['value'] = data['value'].astype(int)

#data.head(10)

# Note :Axis is 0 for rows and 1 for colmns

data.drop('quality',axis=1,inplace=True)

# I think there is no need for data pre processing but still can eleminate null values
data.dropna(inplace= True )




#Specified the features of the models
x = data.iloc[:, :-1]  # Selects all columns except the last one
y = data.iloc[:, -1]   # Selects only the last column



# Lets split into tarining and testing dataset

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

#Scaling of the features:

from sklearn.preprocessing import StandardScaler
std_x=StandardScaler()

x_train=std_x.fit_transform(x_train)
x_test=std_x.transform(x_test)


#Fitting the data into DataClassifier
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(x_train,y_train)

# Testing the data
y_pred= classifier.predict(x_test)


from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

f1 = f1_score(y_test, y_pred)
score = accuracy_score(y_test, y_pred)
print("F1 Score for the Decision Trees is {}".format(f1))
print("Accuracy Score for the Decision Trees is {}".format(score))
```

# **Classification Using KNN algorithm**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


file=pd.read_csv("/content/wind_quality.txt")

data=pd.DataFrame(file)

data['value']=''


for i in range(len(data)):
  if data.at[i,'quality']>=7.0:
    data.at[i,'value']=1
  else:
    data.at[i,'value']=0


#Dont know why, but there were some errors while using for loop, so I repeated the step
data['value'] = data['value'].astype(int)

#data.head(10)

data.drop('quality',axis=1,inplace=True)

#data.info()

# I think there is no need for data pre processing but still can eleminate null values
data.dropna(inplace= True )

x = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Lets split into tarining and testing dataset

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

#Scaling of the features:

from sklearn.preprocessing import StandardScaler
std_x=StandardScaler()

x_train=std_x.fit_transform(x_train)
x_test=std_x.transform(x_test)


from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
classifier.fit(x_train, y_train)


y_pred= classifier.predict(x_test)


from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score


k_values = range(1, 21)
f1_scores = []
accuracy=[]
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    f1_scores.append(f1_score(y_test, y_pred))
    accuracy.append(accuracy_score(y_test, y_pred))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(k_values, f1_scores, marker='o', label='F1 Score')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('F1 Score')
plt.title('F1 Score vs. k')
plt.legend()

# Plot Accuracy vs. k
plt.subplot(1, 2, 2)
plt.plot(k_values, accuracy, marker='o', label='Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k')
plt.legend()

plt.show()
```

# **Classification Using Logistic**(without using sklearn)

```python
import numpy as np
import pandas as pd
import random

def train_test_split(X, y, test_size, random_state):
    if random_state:
        np.random.seed(random_state)

    data_size = len(X)
    test_data_size = int(test_size * data_size)

    # Randomly shuffle the indices
    shuffled_indices = np.random.permutation(data_size)

    # Split the indices into training and testing sets
    test_indices = shuffled_indices[:test_data_size]
    train_indices = shuffled_indices[test_data_size:]

    # Create training and testing sets
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]

    return X_train, X_test, y_train, y_test


class LogisticRegression() :
    def __init__( self, learning_rate, iterations ) :
        self.learning_rate = learning_rate
        self.iterations = iterations

    # Function for model training
    def fit( self, X, Y ) :
        # no_of_training_examples, no_of_features
        self.m, self.n = X.shape
        # weight initialization
        self.W = np.zeros( self.n )
        self.b = 0
        self.X = X
        self.Y = Y

        # gradient descent learning

        for i in range( self.iterations ) :
            self.update_weights()
        return self


    def update_weights( self ) :
        A = 1 / ( 1 + np.exp( - ( self.X.dot( self.W ) + self.b ) ) )

        tmp = ( A - self.Y.T )
        tmp = np.reshape( tmp, self.m )
        dW = np.dot( self.X.T, tmp ) / self.m
        db = np.sum( tmp ) / self.m

        self.W = self.W - self.learning_rate * dW
        self.b = self.b - self.learning_rate * db

        return self

    # Hypothetical function  h( x )

    def predict( self, X ) :
        Z = 1 / ( 1 + np.exp( - ( X.dot( self.W ) + self.b ) ) )
        Y = np.where( Z > 0.5, 1, 0 )
        return Y

def accuracy(test, pred):
    val=0
    for i in range( np.size( test ) ) :
        if test[i] == pred[i] :
          val=val+1

    return val/np.size( test )



df = pd.read_csv( "/content/wind_quality.txt" )
df['good']=df['quality']>=7
df.drop(columns=['quality'])
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1:].values

# Splitting dataset into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size = 0.2, random_state = 0 )

# Model training
model = LogisticRegression( learning_rate = 0.01, iterations = 100 )

model.fit( X_train, Y_train )

# Prediction on test set
Y_pred = model.predict( X_test )

# measure performance


print(accuracy(Y_pred, Y_test))
```
