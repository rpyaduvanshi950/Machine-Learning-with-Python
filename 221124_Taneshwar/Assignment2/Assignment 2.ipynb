## Creating Basic Function for Linear Regresstion


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(y_hat, y):
    l = (y_hat - y) ** 2 / 2
    cost = l.mean()
    return cost
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y)
m=x.shape[0]
```

```python
def hypo(w,x,b):
    return w*x+b
```

```python
def gradient(x,y,w,b,lr,m):
   y_hat=hypo(w,x,b)
   dw = (2.0/m) * np.sum(np.multiply(x,(y_hat-y)))
   db = (2.0/m) * np.sum(y_hat-y)
   w = w -lr* dw
   b = b - lr * db
   return w,b
```
