```python
### Logistic Regression from scratch ###
```

```python
X_train_ = np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test_ = np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
def pred(X, weight):
    z = np.dot(X, weight)
    y_pred = sigmoid(z)
    return (y_pred >= 0.5).astype(int)
```

```python
def init_weig(columns):
    return np.zeros((columns, 1))
```

```python
init_weig = init_weig(X_train_.shape[1])
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def gradient_descent(X, y, weights):


  n = len(y)
  lr = 0.01
  iter = 3000

  for i in range(iter):


    z = np.dot(X, weights)
    y_pred = sigmoid(z)

    error = y_pred - y
    grad = np.dot(X.T, error) / n

    weights -= lr * grad

    return weights
```

```python
weights = gradient_descent(X_train_, y_train.values.reshape(-1, 1), init_weig)
```

```python
reg_pred = pred(X_test_, weights)
```

```python
reg_accuracy = accuracy_score(y_test, reg_pred)
reg_f1 = f1_score(y_test, reg_pred)
```
