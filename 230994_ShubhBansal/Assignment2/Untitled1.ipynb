```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2*x -3 + rng.randn(50)
plt.scatter(x, y)
m=x.shape[0]
```

```python
x_max=np.max((x))
y_max=np.max((y))
x_min=np.min((x))
y_min=np.min((y))
x=(x-x_min)/(x_max-x_min)
y=(y-y_min)/(y_max-y_min)
plt.scatter(x,y)
```

```python
def cost(y_hat, y):
    l = (abs((y-y_hat))**3)/2
    cost = l.mean()
    return cost
```

```python
def hypothesis(x,w,b):
    return w*x+b
```

```python
def gradient_descent(x,y,w,b,learning_rate,m):
   y_hat=hypothesis(x,w,b)
   dw = (3.0/2*m) * np.sum(np.multiply(np.multiply(abs(y-y_hat),(y-y_hat)),-x))
   db = (3.0/2*m) * np.sum(np.multiply((abs(y-y_hat)),(y-y_hat)))
   w = w -learning_rate * dw
   b = b - learning_rate * db
   return w,b
```

```python
b = np.random.randn(1)
w = np.random.randn(1)

learning_rate = 0.0006
num_epochs = 1000000

L=[]
for i in range(num_epochs):
    w, b = gradient_descent(x,y,w,b,learning_rate,m)
    y_hat = hypothesis(x,w,b)
    L.append(cost(y_hat,y))
    if (i%(num_epochs/10))==0:
        plt.plot(x,y,'bo')
        plt.plot(x,y_hat,'g--')
        plt.show()
```

```python
print(f"Obtained value of w: {w}")
print(f"Obtained value of b: {b}")
print(f"Obtained line: y={w}x+{b}")
plt.plot(x,y,'bo')
plt.plot(x,y_hat,'g--')
plt.show()
```

```python
plt.plot(L,'r')
plt.show()
```
