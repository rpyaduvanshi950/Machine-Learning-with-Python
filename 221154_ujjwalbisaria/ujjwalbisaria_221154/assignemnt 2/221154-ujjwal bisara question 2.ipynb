```python
x=np.vstack((np.ones((x.shape[0], )), x.T)).T
x2=np.vstack((np.ones((x2.shape[0], )), x2.T)).T
```

```python
def model(x,y,learningrate,iteration):
    m=y.size
    theta=np.zeros((x.shape[1],1))
    cost_list=[]
    for i in range(iteration):
        
        y_pred=np.dot(x,theta)

        cost=(1/(2*m))*np.sum(np.square(y_pred - y))

        d_theta= (1/m)*np.dot(x.T,y_pred - y)
        theta= theta - learningrate*d_theta

        cost_list.append(cost)

        if(iteration%100==0):
            print(cost)

    return cost_list,theta
```

```python
learnrate=0.008
iteration2=10000

cost,theta=model(x,y,learningrate=learnrate,iteration=iteration2)
```

```python
iterations = list(range(0, iteration2))
plt.plot(iterations,cost)
```

```python
y_pred_scratch=np.dot(x2,theta)
```

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x,y)
```

```python
y_pred_skle=ml.predict(x2)
```

```python
error1=(1/x2.shape[0])*(np.sum(np.abs(y2-y_pred_scratch)))
error2=(1/x2.shape[0])*(np.sum(np.abs(y2-y_pred_skle)))
```

```python
print("percentage accurqcy in scartch is",100-error1*100)
print("percentage acuuracy from sklearn is",100- error2*100)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
# Calculate R2 score
r2_Scatch = r2_score(y2, y_pred_scratch)
r2_skle = r2_score(y2, y_pred_skle)

# Calculate Mean Squared Error (MSE)
mse_srcatch = mean_squared_error(y2,y_pred_scratch)
mse_skle = mean_squared_error(y2,y_pred_skle)
```

```python
print("r2 score from sratch",r2_Scatch)
print("r2 score from skle",r2_skle)
print("mse score from sratch",mse_srcatch)
print("mse score from skle",mse_skle)
```
