```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 2 + 3 * x+ np.random.randn(100, 1)
```

```python
plt.scatter(x, y, label='Generated Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Dataset')
plt.legend()
plt.show()
```

```python
xp=np.c_[np.ones((100,1)),x]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xp, y, test_size=0.15, random_state=0)
```

from linear regression model with cost function =(x-x_pred)^3

```python
def linear_regression_cubed_loss(x, y, learning_te,iteration):
    m, n = X_train.shape
    theta = np.random.randn(n, 1)
    cost=[]

    for i in range(iteration):

        ypred=np.dot(x,theta)

        # gradients = (1/m) * np.dot(x.T, np.sign((ypred)) * (ypred)**2) + 3 * np.sign(theta) * np.abs(theta)**2
        # theta = theta - learning_te * gradients

        gradients = (1/3*m) * np.dot(x.T , (np.sign(ypred - y) * (ypred - y)**2)) + 3 * np.sign(theta) * np.abs(theta)**2
        theta = theta - learning_te * gradients


        cosit = np.mean(np.abs(ypred-y)**3)
        cost.append(cosit)

        if(iteration%100==0):
            print(cosit)

    return theta,cost
```

```python
learning_rate_1=0.00001
iterations_1=50000


theta_1,cost_1=linear_regression_cubed_loss(X_train,y_train,learning_rate_1,iterations_1)
```

```python
iterations = list(range(0,iterations_1))
plt.plot(iterations,cost_1)
```

```python
def polynomial_regression_seventh_loss(x, y, learning_rate,iterations):
    m, n = x.shape
    theta = np.random.randn(n, 1)
    cost=[]

    for i in range(iterations):
        ypred = np.dot(x ,theta)
        gradients = (1/m) *np.dot( x.T , (np.sign(ypred -y) * (ypred-y)**6)) + 7 * np.sign(theta) * np.abs(theta)**6
        theta = theta - learning_rate * gradients

    # Compute the cost (mean seventh power error)
        costo = np.mean(np.abs(ypred-y)**7)
        cost.append(costo)

        if(i%100==0):
    
            print(costo)

    return theta, cost
```
