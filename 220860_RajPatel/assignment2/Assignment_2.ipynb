```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Generate random data
X = 10 * np.random.rand(50)
y = 4 + 3 * X + np.random.randn(50)

# Plot the dataset
plt.scatter(X, y, alpha=0.5)
plt.title("Generated Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
def hypothesis(X,w,b):
    
    return w*X-b
```

```python
# Gradient discent
```

```python
def gradient_descent(X,y,w,b,learning_rate,lossfunction):
    y_hat = hypothesis(X,w,b)
    m = len(y)
    
    dw = (3/m)*np.sum(np.multiply(X,(y-y_hat)**(lossfunction-1)))
    db = (3/m)*np.sum((y-y_hat)**(lossfunction-1))
    
    w = w-learning_rate*(-dw)
    b = b-learning_rate*(-db)
    
    return w,b
    
```

```python
w = np.random.randn(1)
b = np.random.randn(1)

learning_rate = 0.0005
iteration = 10

# for loss function |x-x_hat|^3

lossfunction = 3



for i in range(iteration):
    w,b = gradient_descent(X,y,w,b,learning_rate,lossfunction)
    y_hat =  hypothesis(X,w,b)
    
    plt.plot(X,y,'bo')
    plt.plot(X,y_hat,'g--')
    plt.show()
```
